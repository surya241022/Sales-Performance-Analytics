```sql
-- 1) What is the total sales by category?

SELECT
    p.category,
    format_number(round(SUM(f.sales), 2),2) AS total_sales
FROM fact_sales f
JOIN dim_product p
    ON f.dim_product_key = p.dim_product_key
GROUP BY p.category
ORDER BY total_sales DESC;




-- 2) What is the total sales by sub-category?

select 
    p.sub_category,
    format_number(round(sum(f.sales),2),2) as total_sales
from fact_sales f
join dim_product p
    on f.dim_product_key = p.dim_product_key
group by p.sub_category 
order by total_sales desc;

-- 3) Which category has the highest total sales?

select 
    p.category,
    format_number(round(sum(f.sales),2),2) as total_sales
from fact_sales f
join dim_product p
    on f.dim_product_key = p.dim_product_key
group by p.category 
order by total_sales desc
limit 1;

-- 4) Which sub-category has the highest total sales?

select 
    p.sub_category,
    format_number(round(sum(f.sales),2),2) as total_sales
from fact_sales f
join dim_product p
    on f.dim_product_key = p.dim_product_key
group by p.sub_category 
order by total_sales desc
limit 1;

-- 5) What is the monthly sales trend across all categories?

select 
    date_format(f.order_date, "yyyy-MM") as order_month,
    -- date_format(f.order_date, 'yyyy') as order_year,
    format_number(round(sum(f.sales),2),2) as total_sales
from fact_sales f
group by order_month
order by order_month desc;



-- 6) How does monthly sales trend differ by category?

SELECT
    date_format(f.order_date, 'yyyy-MM') AS sales_month,
    p.category,
    format_number(ROUND(SUM(f.sales), 2), 2) AS total_sales
FROM fact_sales f
JOIN dim_product p
    ON f.dim_product_key = p.dim_product_key
GROUP BY
    date_format(f.order_date, 'yyyy-MM'),
    p.category
ORDER BY
    sales_month,
    SUM(f.sales) DESC;




-- 7) What is the total sales by customer segment?

select
    c.segment,
    format_number(round(sum(f.sales),2),2) as total_sales
from fact_sales f
join dim_customer c
    on f.dim_customer_key = c.dim_customer_key
group by c.segment
order by total_sales desc;

-- 8) Which customer segment contributes the most revenue?

select
    c.segment,
    format_number(round(sum(f.sales),2),2) as total_sales
from fact_sales f
join dim_customer c
    on f.dim_customer_key = c.dim_customer_key
group by c.segment
order by total_sales desc
limit 1;

-- 9) What is the total sales by region?

select
    l.region,
    format_number(round(sum(f.sales),2),2) as total_sales
from fact_sales f
join dim_location l
    on f.dim_location_key = l.dim_location_key
group by l.region
order by total_sales desc;

-- 10) Which region generates the highest sales?

select
    l.region,
    format_number(round(sum(f.sales),2),2) as total_sales
from fact_sales f
join dim_location l
    on f.dim_location_key = l.dim_location_key
group by l.region
order by total_sales desc
limit 1;

-- 11) What is the total sales by state?

select 
    l.state,
    format_number(round(sum(f.sales),2),2) as total_sales
from fact_sales f
join dim_location l
    on f.dim_location_key = l.dim_location_key
group by l.state
order by total_sales desc;

-- 12) Which top 10 customers generate the highest sales?

select 
    l.state,
    format_number(round(sum(f.sales),2),2) as total_sales
from fact_sales f
join dim_location l
    on f.dim_location_key = l.dim_location_key
group by l.state
order by total_sales desc
limit 10;

-- 13) What is the average order value overall?

SELECT
    format_number(
        ROUND(
            SUM(f.sales) / COUNT(DISTINCT f.order_id),
            2
        ),
        2
    ) AS avg_order_value
FROM fact_sales f;

-- 14) What is the average order value by customer segment?

select 
    c.segment,
    format_number(round(sum(f.sales)/COUNT(DISTINCT f.order_id),2),2) as avg_order_value
from fact_sales f
join dim_customer c
    on f.dim_customer_key = c.dim_customer_key
group by c.segment
order by avg_order_value desc;

-- 15) Which categories perform best in each region?

WITH category_region_sales AS (
    SELECT
        l.region,
        p.category,
        SUM(f.sales) AS total_sales
    FROM fact_sales f
    JOIN dim_product p
        ON f.dim_product_key = p.dim_product_key
    JOIN dim_location l
        ON f.dim_location_key = l.dim_location_key
    GROUP BY
        l.region,
        p.category
),
ranked_categories AS (
    SELECT
        region,
        category,
        total_sales,
        ROW_NUMBER() OVER (
            PARTITION BY region
            ORDER BY total_sales DESC
        ) AS rn
    FROM category_region_sales
)
SELECT
    region,
    category,
    format_number(ROUND(total_sales, 2),2) AS total_sales
FROM ranked_categories
WHERE rn = 1
ORDER BY total_sales DESC;




select * from dim_location;


SELECT *
FROM (
    SELECT
        *,
        ROW_NUMBER() OVER (
            PARTITION order_id,
            ORDER BY ) AS rn
    FROM dim_location
) t
WHERE rn > 1;


SELECT
    postal_code,
    city,
    state,
    COUNT(*) AS cnt
FROM dim_location
GROUP BY
    postal_code,
    city,
    state
HAVING COUNT(*) > 1;





















SELECT
    dim_location_key,
    COUNT(*) AS cnt
FROM dim_location
GROUP BY dim_location_key
HAVING COUNT(*) > 1;

select * from dim_location order by dim_location_key;












```
