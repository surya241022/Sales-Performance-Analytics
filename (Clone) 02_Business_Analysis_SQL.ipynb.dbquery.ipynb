```sql
--Duplicates in FACT SALES

SELECT
    order_id,
    product_id,
    COUNT(*) AS cnt
FROM fact_sales
GROUP BY
    order_id,
    product_id
HAVING COUNT(*) > 1;




--Duplicates in dim customer

SELECT
    customer_id,
    COUNT(*) AS cnt
FROM dim_customer
GROUP BY customer_id
HAVING COUNT(*) > 1;





--Surrogate Key Duplicates
SELECT
    dim_customer_key,
    COUNT(*) AS cnt
FROM dim_customer
GROUP BY dim_customer_key
HAVING COUNT(*) > 1;




--Duplicates in dim product

SELECT
    product_id,
    COUNT(*) AS cnt
FROM dim_product
GROUP BY product_id
HAVING COUNT(*) > 1;





--Surrogate Key Duplicates
SELECT
    dim_product_key,
    COUNT(*) AS cnt
FROM dim_product
GROUP BY dim_product_key
HAVING COUNT(*) > 1;





--Duplicates in dim location

SELECT
    postal_code,
    city,
    state,
    COUNT(*) AS cnt
FROM dim_location
GROUP BY
    postal_code,
    city,
    state
HAVING COUNT(*) > 1;




--Surrogate Key Duplicates
SELECT
    dim_location_key,
    COUNT(*) AS cnt
FROM dim_location
GROUP BY dim_location_key
HAVING COUNT(*) > 1;








SELECT COUNT(*) AS unmatched_product_rows
FROM fact_sales
WHERE dim_product_key IS NULL;








SELECT COUNT(*) AS orphan_rows
FROM fact_sales f
LEFT JOIN dim_product p
  ON f.dim_product_key = p.dim_product_key
WHERE p.dim_product_key IS NULL;









SELECT
    COUNT(*) AS blank_category_rows
FROM dim_product
WHERE category IS NULL
   OR TRIM(category) = '';

```
